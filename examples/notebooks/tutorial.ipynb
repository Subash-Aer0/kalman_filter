```python
import numpy as np
from eskf.core import ESKF
from eskf.measurements import IMUMeasurement, GNSSMeasurement
```

## Initialize the Filter

```python
# Create an ESKF instance
ekf = ESKF(state_dim=9, measurement_dim=3)
print(f"Filter initialized with state dimension {ekf.state_dim}")
```

## Run a Simple Simulation

```python
# Simulation parameters
dt = 0.01  # 100 Hz
duration = 5.0
num_steps = int(duration / dt)

# Run simulation
for i in range(num_steps):
    t = i * dt
    u = np.array([0, 0, 9.81])
    ekf.predict(u, dt)

print(f"Simulation complete: {num_steps} steps")
```
